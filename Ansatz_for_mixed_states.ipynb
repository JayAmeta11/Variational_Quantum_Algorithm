```python
import pennylane as qml
from pennylane import numpy as np
```

```python
device = qml.device('default.qubit',wires=2)
```

```python
@qml.qnode(device)
def purification(weights):
    qml.RY(weights[0],wires=0)
    qml.RY(weights[1],wires=1)
    qml.CNOT(wires=[0,1])
    qml.RX(weights[2],wires=0)
    qml.RX(weights[3],wires=1)
    return qml.density_matrix(wires=[0])
```

```python
weights = np.random.uniform(0,np.pi,4)
rho = purification(weights)
print(f'density matrix of state is {np.round(rho,5)}')
```

```python
print(f'trace of density matrix is {np.trace(rho)}')
```

```python
print(f'purity of states is {np.real(np.trace(np.dot(rho,rho)))}')
```
