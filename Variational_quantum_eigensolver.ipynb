```python
import pennylane as qml
from pennylane import numpy as np
```

```python
weightage = [1.0,0.8,1.6]
observable = [qml.PauliZ(0),qml.PauliX(1),qml.PauliZ(0)@qml.PauliZ(1)]
hamil = qml.Hamiltonian(weightage,observable)
```

```python
num = 2
device = qml.device('default.qubit',wires=num)
@qml.qnode(device)
def quantum_circuit(theta):
    qml.RX(theta[0],wires=0)
    qml.RX(theta[1],wires=1)
    qml.CNOT(wires=[0,1])
    qml.RY(theta[2],wires=0)
    qml.RY(theta[3],wires=1)
    return qml.expval(hamil)
```

```python
theta = np.random.uniform(0,10,4,requires_grad = True)
optimiser = qml.GradientDescentOptimizer(stepsize=0.3)
steps = 40
for i in range(steps):
    theta,cost = optimiser.step_and_cost(quantum_circuit,theta)
    if i%5==0:
        print(f'current paramter is {theta} and cost is {cost}')
print('\n')
print(f'final paramter is {theta} and ground_state energy defined by our hamiltonian is {quantum_circuit(theta)}')
```
