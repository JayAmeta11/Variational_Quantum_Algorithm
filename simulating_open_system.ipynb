```python
import pennylane as qml
from pennylane import numpy as np
```

```python
device1 = qml.device('default.mixed',wires=1)
```

```python
@qml.qnode(device1)
def open_system(theta,damping_rate):
    qml.RX(theta[0],wires=0)
    qml.RY(theta[1],wires = 0)
    qml.AmplitudeDamping(damping_rate, wires=0)
    return qml.expval(qml.PauliZ(0))
```

```python
np.random.seed(42)
theta=np.random.uniform(0,np.pi,2,requires_grad = True)
damping_rate = 0.1
optimiser = qml.GradientDescentOptimizer(stepsize=0.2)
for i in range(5):
    theta,cost = optimiser.step_and_cost(lambda p:open_system(p,damping_rate),theta)
print(f'final parameter is {theta}')
```

```python
dev_pure = qml.device("default.qubit", wires=2)
@qml.qnode(dev_pure)
def pure_state_ansatz(params):
    qml.RX(params[0], wires=0)
    qml.RY(params[1], wires=0)
    return qml.expval(qml.PauliZ(0))
M_matrix = qml.metric_tensor(pure_state_ansatz)(theta)
print("\nCalculated Metric Tensor (M matrix):")
print(np.round(M_matrix, 4))
```
